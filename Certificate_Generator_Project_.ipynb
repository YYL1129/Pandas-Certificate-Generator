```python
# Create certificate logic function

# Basic Recap of Function
def addition(num1,num2):
  print(num1 + num2)

addition(3,5)

```

```python
# in this function, we will hv the logic to create a cert
# this function will have 4 parameters (name,courseName,courseLevel,date)
# We are giong to reuse this function 11 times (since there are 11 students) -> by using a loop
# For each students, there is their name, course, level and date

def certificate_generator(name,courseName,courseLevel,date):

  # There will be 11 files. We dont want to have pdf1, pdf2, pdf3 and etc -> this is not clear and helpful, when we want to find specific person pdf file

  # So what we can do is to attach the name,courseName,courseLevel into the naming of the pdf file
  pdfFileName = "/content/drive/MyDrive/Python Certificates/generator-certificates/" + name + "-" + courseName + "-" +  courseLevel + ".pdf"

  # e.g Veshal-Python-Beginner.pdf -> one 1 glance we know the information of the pdf file
  # we have not create the file yet, we just create the name of the file

  # canvas = blank screen (imagine canvas as the art board in art class)

  # create a canvas object from reportLab import and insert our certificate template folder path into it

  # create a variable c
  # pagesize = landscape(A4) -> making the canvas to be landscape A4 size
  c = canvas.Canvas(pdfFileName,pagesize=landscape(A4))
  # we are creating a blank canvas where it is in landscape and its A4 size

  # the canvas has a function call drawImage -> this is use to draw contents in the canvas
  # drawImage needs 5 parameters value

  # 1st --> Image --> what are you going to draw on the canvas??
  # 2nd --> x axis --> on which coordinate in the horizontal line, in what position u want to draw??
  # 3rd --> y axis --> on which coordinate in the vertical line, in what position u want to draw??
  # 4th --> width --> A4 [1] --> this A4 is the variable we imported in the beginning
  # 5th --> height --> A4 [0]

  imagePath = "/content/drive/MyDrive/Python Certificates/certificate_template.jpg"

  c.drawImage(imagePath,0,0,width=A4[1],height=A4[0])

  # I want the size of the image to be a normal A4 width and height

  #width
  center_x = c._pagesize[0] / 2

  #height
  center_y = c._pagesize[1] / 2

  # 1. name
  c.setFont("Lora-Bold",30)
  c.drawCentredString(center_x, center_y - 46, name)


  # 2. CourseName & courseLevel
  c.setFont("Lora-Bold",28)
  c.drawCentredString(center_x, center_y - 105, courseName + " - " + courseLevel)

  # 3. Date
  c.setFont("Lora-Bold",17)
  c.drawCentredString(center_x + 190, center_y - 160, date)

  # 4. Cert ID
  cert_id = "Cert ID:" + str(int(pd.Timestamp.now().timestamp()))
  c.setFont("Lora-Bold",12)
  c.drawCentredString(center_x + 266, center_y - 230, cert_id.upper())

  c.save() # save the canvas

  # up til this point, we basically hv an empty A4 landscape canvas --> drew it with the cert template image


```

```python
# Lets test the function
certificate_generator("Jun","Python","Beginner","20/06/2020")
certificate_generator("Yee","Python","Beginner","12/12/2024")
```

```python
# certificate_generator function needs 4 inputs --> name,courseName,courseLevel,date

# iterrows --> iterable(loopable) rows

for index,row in df.iterrows():
  certificate_generator(row["Name"],row["Course"],row["CourseLevel"],row["FormattedDate"])
```
